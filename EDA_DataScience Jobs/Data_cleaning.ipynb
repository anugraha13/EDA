# Salary Parsing

```python
df['hourly']=df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df['employer_provided']=df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary' in x.lower() else 0)
```

```python
df
```

```python
df=df[df['Salary Estimate']!='-1'] # removing negative salary estimate
df
```

```python
salary=df['Salary Estimate'].apply(lambda x:x.split('(')[0])
salary
```

```python
minus_kd=salary.apply(lambda x:x.replace('$','').replace('K',''))
minus_kd
```

```python
minus_hr=minus_kd.apply(lambda x:x.lower().replace('per hour','').replace('employer provided salary:',''))
minus_hr
```

```python
df
```

```python
df['min_salary']=minus_hr.apply(lambda x:int(x.split('-')[0]))
df['max_salary']=minus_hr.apply(lambda x:int(x.split('-')[1]))
```

```python
df['min_salary'].dtype
```

```python
df['avg_salary']=(df.min_salary+df.max_salary)/2
```

```python
df['avg_salary']
```

# Company name

```python
df.columns
```

```python
df['company_txt']=df.apply(lambda x: x['Company Name'] if x['Rating'] <0 else x['Company Name'][:-3],axis=1)
```

```python
df['company_txt']
```

# State Name

```python
df['job_state']=df['Location'].apply(lambda x : x.split(',')[1])
```

```python
df.job_state.value_counts()
```

# Checking company location and Headquarters are in same place

```python
df['same_state']=df.apply(lambda x: 1 if x.Location==x.Headquarters else 0,axis=1)
```

```python
df['same_state']
```

# Age of Company

```python
df['age']=df.Founded.apply(lambda x: 1 if x<1 else 2022-x)
```

```python
df['age']
```
