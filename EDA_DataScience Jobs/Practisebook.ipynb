# Job Title

```python
def titlesimplifier(title):
    if 'data scientist' in title.lower():
        return 'Data Scientist'
    elif 'analyst' in title.lower():
        return 'Analyst'
    elif 'engineer' in title.lower():
        return 'Data Engineer'
    elif 'machine learning' in title.lower():
        return 'MLE'
    elif 'manager' in title.lower():
        return 'Manager'
    else:
        return 'NA'
    
    
    
```

```python
df['Designation']=df['Job Title'].apply(titlesimplifier)
```

```python
df['Designation'].value_counts()
```

```python
def seniority(title):
    if 'jr' in title.lower() or 'junior' in title.lower():
        return 'Junior'
    elif 'sr' in title.lower() or 'senior' in title.lower():
        return 'Senior'
    else:
        return 'NA'
```

```python
df['Seniority']=df['Job Title'].apply(seniority)
```

```python
df['Seniority'].value_counts()
```

# Salary Estimate

```python
df=df[df['Salary Estimate']!='-1']
```

```python
df[df['Salary Estimate']=='-1']['Job Title']
```

```python
salary=df['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
salary
```

```python
remove_kd=salary.apply(lambda x:x.replace('K','').replace('$',''))
```

```python
remove_kd
```

```python
df['Employer Provided']=df['Salary Estimate'].apply(lambda x :1 if 'employer provided' in x.lower() else 0)
```

```python
df['Per Hour']=df['Salary Estimate'].apply(lambda x :1 if 'per hour' in x.lower() else 0)
```

```python
remove_kd=remove_kd.apply(lambda x : x.replace('Employer Provided Salary:','').replace('Per Hour',''))
```

```python
df['min_salary']=remove_kd.apply(lambda x:int(x.split('-')[0]))
```

```python
df['max_salary']=remove_kd.apply(lambda x:int(x.split('-')[1]))
```

```python
remove_kd.to_csv('check.csv',index=None)
```

```python
df['avg_salary']=(df.min_salary+df.max_salary)/2
```
