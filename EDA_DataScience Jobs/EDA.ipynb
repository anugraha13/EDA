# Feature Engineering

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('salary_data_cleaned.csv')
```

```python
df
```

```python
def title_simplifer(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'mle'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    else:
        return 'NA'
    
```

```python
def seniority(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'sr.' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
        return 'Senior'
    elif 'jr' in title.lower() or 'junior' in title.lower() or 'jr.' in title.lower():
        return 'Junior'
    else:
        return 'NA'
```

```python
#simplify the job title

df['job_simp']=df['Job Title'].apply(title_simplifer)
```

```python
df.job_simp.value_counts()
```

```python
# simplify the seniority

df['seniority']=df['Job Title'].apply(seniority)
```

```python
df.seniority.value_counts()
```

```python
df.columns
```

```python
df.job_state.value_counts()
```

```python
df.job_state
```

```python
# fix the los angeles

df['job_state']=df.job_state.apply(lambda x : x.strip() if x.strip().lower()!='los angeles' else 'CA')
df.job_state.value_counts()
```

```python
# job description length
df['desc_len']=df['Job Description'].apply(lambda x : len(x))
df['desc_len']
```

```python
# competitor count

df['Competitors']
```

```python
df['comp_count']=df['Competitors'].apply(lambda x : len(x.split(',')) if x!='-1' else 0)
```

```python
df['comp_count']
```

```python
df.columns
```

```python
# hourly wage to annual

df['min_salary']=df.apply(lambda x : x.min_salary*2 if x.hourly==1 else x.min_salary,axis=1)
df['max_salary']=df.apply(lambda x : x.max_salary*2 if x.hourly==1 else x.max_salary,axis=1)

```

```python
df[df.hourly==1][['hourly','min_salary','max_salary']]
```

```python
df.company_txt
```

```python
df['company_txt']=df.company_txt.apply(lambda x : x.replace('\n',''))
```
