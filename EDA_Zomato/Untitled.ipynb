```python
df_country=pd.read_excel('Country-code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.columns
```

```python
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index #getting the country name
country_names
```

```python
country_val=final_df.Country.value_counts().values #getting the values
country_val
```

```python
## pie chart - Top 3 countries

plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')

# observation -- Zomato maximum records/transactions are from India, After that, US and UK.
```

# Finding Relationships

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
## Observation - Most of the people are not rated.
# when rating 4.5 to 4.9 -- Excellent , 4.0 to 4.4 -- very Good, 
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6) # to increase the size of the chart
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,hue='Rating color',palette=['white','red','orange','yellow','green','green'])
```

```python
# observation - No one rated 5, More rated 0, maximum number of rating is between 2.5 and 3.4
```

```python
sns.countplot(x="Rating color", data=ratings,palette=['white','red','orange','yellow','green','green']) # this is not counting the rating count.. 
# it is about the frequency
```
